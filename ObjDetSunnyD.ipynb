```python
from ultralytics import YOLO
```

```python
model = YOLO("yolo11n.pt") # load a pretrained model (recommended for training)

#model = YOLO("yolo11n.yaml").load("yolo11n.pt") # build from YAML and transfer weights
```

```python
model.train(data="/content/data.yml", epochs=20)
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Load the CSV file
file_path = 'runs/detect/train52/results.csv'
results_df = pd.read_csv(file_path)

# Extract data for plotting
epochs = results_df['epoch']
train_box_loss = results_df['train/box_loss']
train_cls_loss = results_df['train/cls_loss']
train_dfl_loss = results_df['train/dfl_loss']
val_box_loss = results_df['val/box_loss']
val_cls_loss = results_df['val/cls_loss']
val_dfl_loss = results_df['val/dfl_loss']

# Plot the training and validation losses
plt.figure(figsize=(12, 6))

# Training Losses
plt.plot(epochs, train_box_loss, label="Train Box Loss", linestyle='--')
plt.plot(epochs, train_cls_loss, label="Train Class Loss", linestyle='--')
plt.plot(epochs, train_dfl_loss, label="Train DFL Loss", linestyle='--')

# Validation Losses
plt.plot(epochs, val_box_loss, label="Validation Box Loss", linestyle='-')
plt.plot(epochs, val_cls_loss, label="Validation Class Loss", linestyle='-')
plt.plot(epochs, val_dfl_loss, label="Validation DFL Loss", linestyle='-')

plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.title("Training and Validation Loss Over Epochs")
plt.legend()
plt.grid(visible=True, linestyle='--', alpha=0.5)
plt.show()
```
